# **1. How are the efficiency numbers looking like?**

```python
# Calculate average resolution time and average first response time for each group
average_resolution_time = df.groupby('Group')['Resolution time'].mean()
average_first_response_time = df.groupby('Group')['First reply time in minutes within business hours'].mean()

# Print the results
print("Average Resolution Time by Group:")
print(average_resolution_time)
print("\nAverage First Response Time by Group:")
print(average_first_response_time)
```

Average Resolution Time by Group:

1. The average resolution time for the "Support" group is approximately 116.51 minutes.
This indicates that, on average, it takes the Support group about 116.51 minutes to resolve a ticket.
Average First Response Time by Group:

2. The average first response time for the "Support" group is approximately 307.51 minutes.

This indicates that, on average, it takes the Support group about 307.51 minutes to send the first response to a ticket within business hours.
In summary, the efficiency numbers suggest that while the Support group has a relatively quick average resolution time, the average first response time is higher, indicating potential areas for improvement in terms of response speed.

## **3. What type of tickets are taking the most time to resolve? Create a table + relevant charts. Create the different type of data types we can infer from this data**

```python
import matplotlib.pyplot as plt

# Calculate average resolution time for each ticket type
average_resolution_time_by_ticket_type = df.groupby('Manual Tagging of Categories [list]')['Resolution time'].mean()

# Sort ticket types by average resolution time
average_resolution_time_by_ticket_type = average_resolution_time_by_ticket_type.sort_values(ascending=False)

# Print the table
print("Average Resolution Time by Ticket Type:")
print(average_resolution_time_by_ticket_type)

# Plot the bar chart
plt.figure(figsize=(10, 6))
average_resolution_time_by_ticket_type.plot(kind='bar', color='salmon')
plt.xlabel('Ticket Type')
plt.ylabel('Average Resolution Time (minutes)')
plt.title('Average Resolution Time by Ticket Type')
plt.grid(axis='y')
plt.show()
```
